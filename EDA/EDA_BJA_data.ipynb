```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
BJA = pd.read_csv(
    "https://raw.githubusercontent.com/Ramil-cyber/Research_Linking_Analyzing_Deaths_US_Prisons/refs/heads/main/Data/BJA_updated_fields_Full-Data_data-882_Cleaned-up.2.25.csv"
)
BJA
```

```python
missing_counts = BJA.isnull().sum()
missing_percent = (missing_counts / len(BJA)) * 100

missing_summary = pd.DataFrame(
    {"Missing Count": missing_counts, "Missing Percentage (%)": missing_percent}
)

print(missing_summary)
```

```python
# Dropping the columns with missing values
dropped_cols = [
    "Masked Count",
    "Notes",
    "Status - Please use this column to track any updates you can make to these records.",
    "Flag",
    "Flag Count",
    "Gender (Open Text)",
    "Other (Open Text)",
    "Recode",
    "Agency/End Date",
]
BJA.drop(columns=dropped_cols, inplace=True)
```

```python
BJA.head()
```

```python
# Filling missing values and preprocessing

median_age = BJA["Age"].median()
BJA["Age"].fillna(median_age, inplace=True)
BJA["Agency Name"].fillna("Unknown Agency", inplace=True)
BJA["Decedent Last Name"].fillna("Unknown", inplace=True)
BJA["First Name"].fillna("Unknown", inplace=True)
BJA["City"].fillna("Unknown City", inplace=True)
BJA["Zip Code"].fillna("00000", inplace=True)
BJA["Location of Death"].fillna("Unknown Location", inplace=True)
BJA["Street Address:"].fillna("Unknown Location", inplace=True)
BJA["Location Type"] = BJA["Location Type"].replace({"jail": "Jail"})
```

### 1.3 Distribution of “Manner of Death”

```python
# Computiing manner counts short as before
manner_counts_short = BJA[BJA["Year of Fiscal Year of Death"].between(2020, 2023)][
    "Manner of Death"
].value_counts()

# Defining a mapping from long labels - short labels
short_label_map = {
    "Natural causes": "Natural",
    "Unavailable: investigation pending": "Pending",
    "Suicide": "Suicide",
    "Death attributed to use of force by a law enforcement or corrections officer": "Force‑related",
    "Accident": "Accident",
    "Homicide (e.g., an incident between two or more incarcerated individuals resulting in a death)": "Homicide",
    "Other": "Other",
    "Execution": "Execution",
}

# If there happen to be any other categories not in the map, they will stay as is:
labels_long = manner_counts_short.index.tolist()
labels_short = [short_label_map.get(lbl, lbl) for lbl in labels_long]

# Ploting with horizontal bars or vertical bars, but use shortened x‑labels
plt.figure(figsize=(8, 6), dpi=350)

ax = manner_counts_short.plot(kind="bar", color="#66c266", edgecolor="black")

# Title and axis labels
ax.set_title("Manner of Death between 2020 & 2023", fontsize=14)
ax.set_xlabel("Manner of Death", fontsize=12)
ax.set_ylabel("Number of Prisoners", fontsize=12)

# Replacing the verbose x‑ticks with our shorter versions
ax.set_xticklabels(labels_short, rotation=45, ha="right", fontsize=12)
ax.tick_params(axis="y", labelsize=12)

# Adding horizontal gridlines with low visibility
ax.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.3)
plt.tight_layout()
plt.show()
```

### 2.1 Gender & Race/Ethnicity

```python
# Geting race counts for male and female separately
male_counts = BJA[BJA["Gender"] == "Male"]["Race"].value_counts().sort_index()
female_counts = BJA[BJA["Gender"] == "Female"]["Race"].value_counts().sort_index()

# Ensuring both series include all race categories (fill missing with 0)
all_races = sorted(BJA["Race"].dropna().unique())
male_counts = male_counts.reindex(all_races, fill_value=0)
female_counts = female_counts.reindex(all_races, fill_value=0)

# Defining a color palette
custom_colors = [
    "#1f77b4",
    "#ff7f0e",
    "#2ca02c",
    "#d62728",
    "#9467bd",
    "#8c564b",
    "#e377c2",
]

# Creating a figure with three columns: left pie, middle legend, right pie.
fig, axes = plt.subplots(
    1,
    3,
    figsize=(23, 10),
    dpi=350,
    gridspec_kw={"width_ratios": [2, 1, 2], "wspace": 0.4},
)

ax_male, ax_legend, ax_female = axes


# Customing autopct: only show percentages ≥ 2%
def autopct_threshold(pct, threshold=2.0):
    return f"{pct:.1f}%" if pct >= threshold else ""


# Creating Male donut chart
wedges_m, _, autotexts_m = ax_male.pie(
    male_counts,
    labels=None,
    autopct=lambda pct: autopct_threshold(pct, threshold=2.0),
    pctdistance=0.75,
    startangle=90,
    colors=custom_colors,
    wedgeprops={"edgecolor": "white", "linewidth": 1.5},
    textprops={"fontsize": 20, "color": "black"},
)

# Adding white circle to create donut shape
centre_circle_m = plt.Circle((0, 0), 0.50, color="white")
ax_male.add_artist(centre_circle_m)

ax_male.set_title(
    "Male: Distribution of Deaths by Race",
    fontsize=28,
    pad=12,
)
ax_male.axis("equal")

for autotext in autotexts_m:
    autotext.set_fontsize(26)

# Creating Female donut chart
wedges_f, _, autotexts_f = ax_female.pie(
    female_counts,
    labels=None,
    autopct=lambda pct: autopct_threshold(pct, threshold=2.0),
    pctdistance=0.75,
    startangle=90,
    colors=custom_colors,
    wedgeprops={"edgecolor": "white", "linewidth": 1.5},
    textprops={"fontsize": 20, "color": "black"},
)

centre_circle_f = plt.Circle((0, 0), 0.50, color="white")
ax_female.add_artist(centre_circle_f)

ax_female.set_title("Female: Distribution of Deaths by Race", fontsize=28, pad=12)
ax_female.axis("equal")

for autotext in autotexts_f:
    autotext.set_fontsize(26)

# Legends in the middle
ax_legend.axis("off")

# Drawing a single legend centered in the middle subplot
legend = ax_legend.legend(
    handles=wedges_m,
    labels=all_races,
    title="Race",
    loc="center",
    fontsize=22,
    title_fontsize=24,
    frameon=True,
    edgecolor="black",
)

# Adjusting handle length and spacing for clarity:
legend._legend_box.align = "left"
for text in legend.get_texts():
    text.set_fontsize(20)

plt.show()
```

```python
# Recomputing counts
race_counts = BJA["Race"].value_counts().sort_values(ascending=False)
eth_counts = BJA["Ethnicity"].value_counts().sort_values(ascending=False)

fig, (ax_race, ax_eth) = plt.subplots(
    1, 2, figsize=(18, 8), gridspec_kw={"width_ratios": [3, 2]}, constrained_layout=True
)

# Race: Bars in a slightly darker lavender
categories_r = race_counts.index.tolist()
values_r = race_counts.values
x_r = np.arange(len(categories_r))

ax_race.bar(x_r, values_r, color="#8a82e8", edgecolor="black")  # darker lavender

ax_race.set_xticks(x_r)
ax_race.set_xticklabels(
    categories_r, rotation=30, ha="right", fontsize=18  # larger than original
)
ax_race.set_ylabel("Count of Deaths", fontsize=18)
ax_race.set_title("Total Prison Deaths by Race", fontsize=18, pad=12)
ax_race.tick_params(axis="y", labelsize=18)

# Griding lines for readability
ax_race.grid(axis="y", linestyle="--", alpha=0.4)


# Ethnicity: Bars in yellow
categories_e = eth_counts.index.tolist()
values_e = eth_counts.values
x_e = np.arange(len(categories_e))

ax_eth.bar(x_e, values_e, color="gold", edgecolor="black")
ax_eth.set_xticks(x_e)
ax_eth.set_xticklabels(categories_e, rotation=30, ha="right", fontsize=18)
ax_eth.set_ylabel("Count of Deaths", fontsize=18)
ax_eth.set_title("Total Prison Deaths by Ethnicity", fontsize=18, pad=12)
ax_eth.tick_params(axis="y", labelsize=18)

# Griding lines for readability
ax_eth.grid(axis="y", linestyle="--", alpha=0.4)
plt.show()
```

### 2.2 Age Distribution & 2.3 Deaths by Location Type

```python
# HISTOGRAM WITHOUT THE OVERLAID LINE
ages = BJA["Age"].dropna()
counts, bin_edges = np.histogram(ages, bins=20)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

plt.figure(figsize=(8, 6), dpi=350)

# Drawing only the histogram bars (no line)
plt.bar(
    bin_centers,
    counts,
    width=(bin_edges[1] - bin_edges[0]) * 0.95,
    color="crimson",
    edgecolor="black",
    alpha=0.8,
)

plt.title("Age Distribution of Decedents", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Number of Decedents", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()


# BAR CHART OF 'Location Type' WITH DARK CYAN
loc_type_counts = BJA["Location Type"].value_counts()

plt.figure(figsize=(8, 6), dpi=350)
loc_type_counts.plot(kind="bar", color="darkcyan", edgecolor="black")

plt.title("Number of Deaths by Location", fontsize=14)
plt.xlabel("Location", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

### 3.3 Year‑over‑Year Change

```python
# Recomputing yearly counts & percent change from BJA
BJA["Death Year"] = pd.to_datetime(BJA["Fiscal Year of Death"]).dt.year
year_counts = BJA["Death Year"].value_counts().sort_index()

yearly = year_counts.to_frame(name="Deaths")
yearly["Pct Change"] = yearly["Deaths"].pct_change() * 100

# Preparing x‑positions and labels
years = yearly.index.astype(int).tolist()
x_vals = np.arange(len(years))
year_labels = [str(y) for y in years]

# Creating a single‐axis figure
fig, ax1 = plt.subplots(figsize=(8, 5), dpi=350)

# Ploting Deaths as bars
bars = ax1.bar(x_vals, yearly["Deaths"], color="steelblue", edgecolor="black")
ax1.set_xticks(x_vals)
ax1.set_xticklabels(year_labels, fontsize=14)
ax1.set_xlabel("Fiscal Year", fontsize=14)
ax1.set_ylabel("Number of Deaths", fontsize=14)
ax1.tick_params(axis="y", labelsize=14)
ax1.set_ylim(0, 8000)

# Annotating each bar with its Death count and Pct Change
base_offset = 8000 * 0.02
pct_offset = 8000 * 0.04

for i, (death_count, pct) in enumerate(zip(yearly["Deaths"], yearly["Pct Change"])):
    ax1.text(
        i,
        death_count + base_offset,
        f"{int(death_count)}",
        ha="center",
        va="bottom",
        fontsize=12,
        color="black",
    )

    # Percent Change annotation
    if not np.isnan(pct):
        pct_color = "green" if pct > 0 else "red"
        ax1.text(
            i,
            death_count + base_offset + pct_offset,
            f"{pct:.1f}%",
            ha="center",
            va="bottom",
            fontsize=11,
            color=pct_color,
        )

# Adding horizontal grid lines on the Deaths axis for readability
ax1.yaxis.grid(True, linestyle=":", linewidth=0.6, alpha=0.7)

plt.title("Prison Deaths by Year with % Change", fontsize=14)
plt.tight_layout()
plt.show()
```

### 4.2 State‑Level Bar Chart

```python
# Computing counts by state and select top 10
state_counts = BJA["Location of Death-State (postal abbreviation):"].value_counts()
top10_states = state_counts.head(10)

# Computing percentages of total deaths for these top 10 states
total_deaths = state_counts.sum()
top10_pct = (top10_states / total_deaths) * 100

# To have the largest‐percentage (highest‐count) bar appear at the top of the chart,
# we sort ascending so that barh plots the smallest on the bottom and the largest on top.
top10_states_sorted = top10_states.sort_values(ascending=True)
top10_pct_sorted = top10_pct[top10_states_sorted.index]

single_color = "sienna"

fig, ax = plt.subplots(figsize=(8, 5), dpi=350)

# Ploting horizontal bars in ascending order
bars = ax.barh(
    top10_states_sorted.index,
    top10_states_sorted.values,
    color=single_color,
    edgecolor="black",
)

# Adding title and axis labels
ax.set_title("Top 10 States by Number of Deaths", fontsize=14)
ax.set_xlabel("Number of Deaths", fontsize=14)
ax.set_ylabel("State", fontsize=14)

# Seting x‐axis limit up to 4000
ax.set_xlim(0, 4000)

# Annotating each bar with its percentage
offset = 4000 * 0.01  # small offset to position text just to the right of each bar
for i, state in enumerate(top10_states_sorted.index):
    count = top10_states_sorted[state]
    pct = top10_pct_sorted[state]
    ax.text(
        count + offset,
        i,
        f"{pct:.1f}%",
        ha="left",
        va="center",
        fontsize=14,
        color="black",
    )

plt.tight_layout()
plt.show()
```

### 6.1 Duplicate Detection

```python
# Checking if “Concat ID” duplicates exist
duplicate_ids = BJA["Concat ID"].duplicated().sum()
print(f"Number of duplicate Concat IDs: {duplicate_ids}")

# Checking composite duplicates by Name + Last Name + Date of Death
composite_dup = BJA.duplicated(
    subset=["First Name", "Decedent Last Name", "Date of Death"]
).sum()
print(f"Number of duplicates by Name + Date of Death: {composite_dup}")
```

```python
# Identifying all Concat IDs that occur more than once
dup_ids = BJA.loc[BJA["Concat ID"].duplicated(keep=False), "Concat ID"].unique()

# Showing every row that has one of those duplicate Concat IDs
duplicate_rows = BJA[BJA["Concat ID"].isin(dup_ids)].sort_values("Concat ID")

duplicate_rows
```

```python
# Finding every row where Name + Date of Death is duplicated
dup_name_date_rows = BJA[
    BJA.duplicated(
        subset=["First Name", "Decedent Last Name", "Date of Death"], keep=False
    )
].sort_values(["First Name", "Decedent Last Name", "Date of Death"])

dup_name_date_rows
```
