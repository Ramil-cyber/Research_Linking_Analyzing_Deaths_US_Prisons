```python
import pandas as pd

jail = pd.read_csv("https://raw.githubusercontent.com/Ramil-cyber/Research_Linking_Analyzing_Deaths_US_Prisons/refs/heads/main/Data/mcijail2016_20192025.csv")
jail
```

```python
prison = pd.read_csv("https://raw.githubusercontent.com/Ramil-cyber/Research_Linking_Analyzing_Deaths_US_Prisons/refs/heads/main/Data/mciprison2016_20192025.csv")
prison
```

```python
import pandas as pd

# Concatenating, aligning on column names and filling missing columns with NaN
merged_df = pd.concat([jail, prison], ignore_index=True, sort=False)

# Display results
print(f"Jail shape: {jail.shape}")
print(f"Prison shape: {prison.shape}")
print(f"Merged shape: {merged_df.shape}")

# Save the merged CSV
merged_df.to_csv('merged_prison_jail_2016_20192025.csv', index=False)
merged_df
```
